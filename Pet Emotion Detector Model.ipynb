```python
import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import librosa
import librosa.display
```

```python
filename = "Mujer Atacada por Dogo Argentino-3xCWI_22Z9A.wav"
y, sr = librosa.load(filename)
# trim silent edges
dogBark, _ = librosa.effects.trim(y)
librosa.display.waveplot(dogBark, sr=sr);
```

```python
n_fft = 2048
hop_length = 512
D = np.abs(librosa.stft(dogBark, n_fft=n_fft,  
                        hop_length=hop_length))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='linear');
plt.colorbar();
```

```python
DB = librosa.amplitude_to_db(D, ref=np.max)
librosa.display.specshow(DB, sr=sr, hop_length=hop_length, 
                         x_axis='time', y_axis='log');
plt.colorbar(format='%+2.0f dB');
```
